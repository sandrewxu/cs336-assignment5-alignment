```python
all_results = []
with jsonlines.open("../results/zero_shot/qwen_gsm8k_zero_shot_20251212_142803.jsonl", "r") as f:
    for result in f:
        all_results.append(result)
```

```python
results_only = all_results[1:]
```

```python
results_only[0]
```

```python
format_and_answer = 0
format_no_answer = 0
no_format_answer = 0
no_format_no_answer = 0
for result in results_only:
    if result['reward'] > 0:
        format_and_answer += 1
    elif result['format_reward'] > 0:
        format_no_answer += 1
    elif result['answer_reward'] > 0:
        no_format_answer += 1
    else:
        no_format_no_answer += 1
```
